```python
import pandas as pd
import numpy as np
```

```python
pokemon=pd.read_csv('Pokemon.csv')
```

```python
pokemon.head(5)
```

```python
# Read one or multiple columns
pokemon[['Name','HP']]
```

```python
# Reading Rows
pokemon.iloc[1:5] # iloc: integer location
```

```python
# Reading specific position
pokemon.iloc[1,1]
```

```python
# iterating rows 
for index,row in pokemon.iterrows():  # iterrows iterate through all the rows and its values to access 
    print(row['Name'])                 #specific column write column name 
```

```python
# Accessing rows based on some criteria 
pokemon.loc[pokemon['Type 1']=='Fire'] # it says access only the rows whose column type1=fire
```

```python
#Soritng values
pokemon.sort_values(['Name','HP'],ascending=[1,0]) # 1: true , 0: false
```

### Making Changes To the Data

```python
#pokemon['Total']=pokemon['HP']+pokemon['Attack']+pokemon['Defense']+pokemon['Sp. Atk']+pokemon['Sp. Def']+pokemon['Speed']+pokemon['Generation']
pokemon['Total']=pokemon.iloc[:,4:10].sum(axis=1) # professional way of doing this
```

```python
pokemon.head(5)
```

```python
pokemon=pokemon.drop(['Total'],axis=1)
```

```python
pokemon.head(3)
```

### Filtering Data

```python
pokemon.loc[(pokemon['Type 1']=='Grass') & (pokemon['Type 2']=='Poison') & (pokemon['HP']>70)] # \ for or
```

```python
pokemon['Name'] # property DataFrame.loc Access a group of rows and columns by label(s) 
```

```python
pokemon.loc[~pokemon['Name'].str.contains('Mega')] # ~ means except
```

```python
import re
pokemon.loc[pokemon['Name'].str.contains('^pi[a-z]*',flags=re.I , regex=True)] # pikachu etc
```

# Conditional Changes

```python
pokemon.loc[pokemon['Type 1']=='Fire','Legendary']=True # if type 1 =fire then legendary = true
```

```python
pokemon.loc[pokemon['Type 1']=='Fire']
```

# Aggregate statistics (groupby)

```python
pokemon.groupby(['Type 1']).mean().sort_values('Defense',ascending=False) # Steel pokemons have the highest defense
            # we can't groupby by Name , because they are distinct
```

```python
pokemon.groupby(['Type 1']).count() # count the similar type 
```
